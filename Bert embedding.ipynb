```python
from transformers import AutoTokenizer, AutoModel

MODEL_NAME = "bert-base-uncased" #"sberbank-ai/sbert_large_mt_nlu_ru"
model = AutoModel.from_pretrained(MODEL_NAME)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
plt.subplot(2, 1, 1)
plt.imshow(model.embeddings.position_embeddings.weight[:100,:].detach().numpy(), cmap='inferno', aspect='auto')
plt.colorbar()
plt.subplot(2, 1, 2)
plt.imshow(-model.embeddings.word_embeddings.weight[::100,:].detach().numpy(), cmap='inferno', aspect='auto')
plt.colorbar()
```
